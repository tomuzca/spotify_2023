```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
spotify_2023 = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')
```

```python
spotify_2023.info()
spotify_2023.columns
```

```python
columns_to_plot = ["danceability_%", "valence_%","energy_%"]

for column in columns_to_plot:
    plt.figure(figsize=(15, 6))
    plt.bar(spotify_2023["released_year"], spotify_2023[column], alpha=0.5)
    plt.title(f"{column} vs released_year")
    plt.xlabel("Released Year")
    plt.ylabel(column)
    plt.grid(True)
    plt.savefig(f'{column}_vs_released_year.png')
    plt.show()
```

```python
for column in columns_to_plot:
   
    decades = np.floor(spotify_2023["released_year"] / 10) * 10
    grouped_data = spotify_2023.groupby(decades)[column].mean()

  
    plt.bar(grouped_data.index, grouped_data.values, width=8, alpha=0.7)  
    plt.title(f"{column} vs released_year (Average per Decade)")
    plt.xlabel("Released Decade")
    plt.ylabel(column)
    plt.grid(axis='y')  
    plt.show()
```
