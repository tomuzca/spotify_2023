```python
#Cargamos el dataset
import pandas as pd 
spotify_2023 = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')
spotify_2023.head()
```

```python
#Info y Columnas para ver por donde empezamos
spotify_2023.info()
spotify_2023.columns
```

```python
#borramos las fechas, por ahora
spotify_2023 = spotify_2023.drop(['released_month', 'released_day', 'released_year'], axis=1)
```

```python
#juntamos a las canciones por artistas
canciones_por_artista = spotify_2023['artist(s)_name'].value_counts()
canciones_por_artista.head(20)
```

```python
#ordenamos por streams
spotify_2023_streams = spotify_2023.sort_values(by='streams', ascending=False)
spotify_2023_streams.head(20)
```

```python
#La cancion Loves Grows de Edison Lighthouse, tiene un error en la canitdad de streams
#Va a ser removida del top 20 para fines de este analisis.

#sin embargo puede que legitimanete forme parte de este DS ya que a pesar de ser una 
#cancion de 1970 ya que es el main theme de SEE YOU IN VENUS, una serie de Netflix del 2023

#mostramos top20 canciones más streameadas 
df_clean1= spotify_2023_streams[spotify_2023_streams['artist(s)_name'] != 'Edison Lighthouse']
df_clean1.head(20)
```

```python
#Juntamos todos los streams por artista
streams_por_artista = df_clean1.groupby('artist(s)_name')['streams'].sum()

# Ordenar los resultados de mayor a menor
streams_por_artista = streams_por_artista.sort_values(ascending=False)

# Mostrar los primeros 10 resultados
streams_por_artista.head(10)
```
