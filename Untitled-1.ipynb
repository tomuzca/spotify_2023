```python
#Cargamos el dataset
import pandas as pd 
spotify_2023 = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')
spotify_2023.head()
```

```python
#Info y Columnas para ver por donde empezamos
spotify_2023.info()
spotify_2023.columns
```

```python
#borramos las fechas, por ahora
spotify_2023 = spotify_2023.drop(['released_month', 'released_day'], axis=1)
```

```python
#juntamos a las canciones por artistas
#top20
canciones_por_artista = spotify_2023['artist(s)_name'].value_counts()
canciones_por_artista.head(20)
```

```python
#La columna streams a pesar de ser números aparecen como object, la cambiamos a int64

spotify_2023['streams'] = pd.to_numeric(spotify_2023['streams'], errors='coerce').astype('Int64')
```

```python
#ordenamos por streams
#hacemos top 20
spotify_2023_streams = spotify_2023.sort_values(by='streams', ascending=False)
stream_top = spotify_2023_streams.head(20)
stream_top.head()
```

```python
import matplotlib.pyplot as plt

# Crear el gráfico de barras
plt.figure(figsize=(15, 6))
canciones_por_artista.plot(kind='bar', color='skyblue')
plt.title('Top 20 Artistas con Más Canciones')
plt.xlabel('Artista(s)')
plt.ylabel('Canciones en top1000')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Juntamos todos los streams por artista
streams_por_artista = spotify_2023.groupby('artist(s)_name')['streams'].sum()

# Ordenar los resultados de mayor a menor
streams_por_artista = streams_por_artista.sort_values(ascending=False)

# Mostrar los primeros 10 resultados
streams_por_artista.head(10)
```

```python
top_20_stream= streams_por_artista.head(20)
plt.figure(figsize=(15, 6))
top_20_stream.plot(kind='bar', color='skyblue')
plt.title('Streamings por artista')
plt.xlabel('Artista(s)')
plt.ylabel('Streams')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
spotify_2023_playlists = spotify_2023.sort_values(by='in_spotify_playlists', ascending=False)
spotify_2023_playlists.head(20)
```

```python
import seaborn as sns
selected_columns = ["streams", "danceability_%", "valence_%", "energy_%", "acousticness_%", "instrumentalness_%", "liveness_%", "speechiness_%"]
subset_df = spotify_2023[selected_columns]

# Calcular la matriz de correlación
correlation_matrix = subset_df.corr()

# Crear el heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5)
plt.title('Correlación entre Streams y Características de la Canción')
plt.show()
```

```python
spotimetrics = ["streams", "danceability_%", "valence_%", "energy_%", "acousticness_%", "instrumentalness_%", "liveness_%", "speechiness_%", "track_name"]
subset_df = stream_top[spotimetrics]

# Establecer 'artist' como índice para facilitar la visualización
subset_df.set_index('track_name', inplace=True)

# Ordenar el DataFrame por cada columna de manera ascendente
subset_df.sort_values(by=subset_df.columns.tolist(), inplace=True)

# Crear un gráfico de barras apiladas
for column in subset_df.columns:
    plt.figure(figsize=(10, 6))
    subset_df[[column]].plot(kind='bar', stacked=True)
    plt.title(f'{column} por Artista')
    plt.xlabel('Artista')
    plt.ylabel('Porcentaje')
    plt.legend().set_visible(False)  # Ocultar la leyenda para mejorar la visualización
    plt.show()
```
